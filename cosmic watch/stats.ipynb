```python
import xml.etree.ElementTree as ET
from pathlib import Path
```

```python
geodata = {}
kml_file_path = Path('f'+str(run)+'/tracklog.kml')
tree = ET.parse(kml_file_path)
root = tree.getroot()
for child in root[0][5][2]:
    tag = child.tag
    key = tag[tag[1:].find('}')+2:]
    if key in geodata:
        geodata[key].append(child.text)
    else:
        geodata[key] = [child.text]
# print(len(geodata['coord']))
```

```python
from datetime import timedelta
```

```python
times = df[['Time']].to_numpy()
start = 0
cutoff = 0
original = geodata['when'][0].split('T')[1].replace('Z','').split(':')
objorigin = timedelta(hours=int(original[0]), minutes=int(original[1]), seconds=int(float(original[2]))) - timedelta(hours=4, seconds=8)
strorigin = str(objorigin)
origin = int(strorigin.replace(':',''))
adjorigin = strorigin.split(':')
for i in range(len(times)):
    tf = int(times[i][0].split(' ')[0].replace(':',''))
    if start == 0:
        if tf > origin:
            print(i, origin, tf)
            start = i
    else:
        if tf > float(geodata['when'][-1].split('T')[1].replace(':','').replace('Z','')) - 40000 - 8:
            cutoff = i
            print(i, tf)
            break
# print(start, cutoff)
timestamp = df[start:cutoff][['TimeStamp [ms]']].to_numpy()
coin = df[start:cutoff][['Coincident']].to_numpy()
valid = np.nonzero(coin)
times = [timestamp[i][0]/1000 for i in valid[0]]
```

```python
nbin = 45
# # wbin = (times[-1]-times[0])/nbin
# # malt = [0 for i in range(nbin)]
# # cuts = [int(str(timedelta(hours=int(adjorigin[0]),minutes=int(adjorigin[1]),seconds=int(adjorigin[2])) + timedelta(seconds=int(wbin*i))).replace(':','')) for i in range(nbin)]
# # cuts.append(int(float(geodata['when'][-1].split('T')[1].replace('Z','').replace(':',''))))
# # # print(cuts)
# # i = 0
# # c = 0
# # for j in range(len(geodata['coord'])):
# #     tj = geodata['when'][j].split('T')[1].replace('Z','').split(':')
# #     tf = float(str(timedelta(hours=int(tj[0]), minutes=int(tj[1]), seconds=float(tj[2])) - timedelta(hours=4, seconds=8)).replace(':',''))
# #     if tf > cuts[i+1]:
# #         malt[i] /= c
# #         i += 1
# #         c = 0
# #     if float(geodata['coord'][j].split(' ')[-1]) > 0:
# #         malt[i] += float(geodata['coord'][j].split(' ')[-1])  
# #         c += 1 
# # malt[-1] /= c
# # xbin = [times[0] + wbin//2 + wbin*i for i in range(nbin)]
# # print(malt)
# # print(xbin)

wbin = (times[-1]-times[0])/nbin
print(wbin)
sumdead = [0 for i in range(nbin)]
i = 0
for j in range(start, cutoff):
    if (timestamp[j-start]/1000-times[0]) > (i+1)*wbin:
        i += 1
    if dead[j] > 0 and coin[j-start] > 0:
        sumdead[i] += dead[j][0]/1e6
print(np.max(sumdead))  
xbin = [times[0] + wbin//2 + wbin*i for i in range(nbin)]

dt = []
for t in geodata['when']:
    ts = t.split('T')[1].replace('Z','').split(':')
    dt.append(times[0] + float((timedelta(hours=int(ts[0]), minutes=int(ts[1]), seconds=float(ts[2])) - timedelta(hours=4, seconds=8) - objorigin).total_seconds()))
alt = []
for l in geodata['coord']:
    a = float(l.split(' ')[-1])
    if a < 0:
        alt.append(0)
    else:
        alt.append(a)
valid = np.nonzero(alt)
dt = [dt[j] for j in valid[0]]
alt = [alt[j] for j in valid[0]]

ps = df[start:cutoff][['Presure [Pa]']].to_numpy()
pt = df[start:cutoff][['TimeStamp [ms]']].to_numpy()/1000

plt.clf()
fig, ax = plt.subplots()
axa = ax.twinx()
fig.set_size_inches(12, 4)
fig.set_dpi(100)
ax.set_xlabel('Timestamp (s)')
ax.set_ylabel('Count')
# ax.hist(timestamp/1000, bins=nbin, edgecolor='black', linewidth=1.2, histtype='step')
# binned = ax.hist(times, bins=nbin, edgecolor='black', linewidth=1.2, histtype='step')
binned = ax.hist(times, bins=nbin, color=[0,0,0,0.2], linewidth=1.2)
# ax.errorbar(xbin, binned[0], yerr=np.sqrt(binned[0]), fmt='none', ecolor='black', capsize=0)
axa.set_ylabel('Altitude (m)')
axa.plot(dt, alt, 'r--')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
# plt.savefig('plots/countaltitude.png')
# print(binned)
plt.show()

plt.clf()
fig, ax = plt.subplots()
axa = ax.twinx()
fig.set_size_inches(12, 4)
fig.set_dpi(100)
ax.set_xlabel('Timestamp (s)')
ax.set_ylabel('Rate (Hz)')
b = [binned[0][i]/(wbin - sumdead[i]) for i in range(nbin)]
berr = [np.sqrt(binned[0][i])/(wbin - sumdead[i]) for i in range(nbin)]
ax.plot(xbin, b)
ax.fill_between(xbin, [b[i]-berr[i] for i in range(nbin)], [b[i]+berr[i] for i in range(nbin)], color=[0.5,0.5,0.5], alpha=0.25, linewidth=0)
# ax.errorbar(xbin, b, yerr=berr, color='black', fmt='.', ecolor='black', capsize=3, elinewidth=1.8)
axa.set_ylabel('Altitude (m)')
axa.plot(dt, alt, 'r--')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
# plt.savefig('plots/ratealtitude.png')
# print(binned)
plt.show()

plt.clf()
fig, ax = plt.subplots()
axa = ax.twinx()
fig.set_size_inches(12, 4)
fig.set_dpi(100)
ax.set_xlabel('Timestamp (s)')
ax.set_ylabel('Rate (Hz)')
b = [binned[0][i]/(wbin - sumdead[i]) for i in range(nbin)]
berr = [np.sqrt(binned[0][i])/(wbin - sumdead[i]) for i in range(nbin)]
ax.plot(xbin, b)
ax.fill_between(xbin, [b[i]-berr[i] for i in range(nbin)], [b[i]+berr[i] for i in range(nbin)], color=[0.5,0.5,0.5], alpha=0.25, linewidth=0)
# ax.errorbar(xbin, b, yerr=berr, color='black', fmt='.', ecolor='black', capsize=3, elinewidth=1.8)
axa.set_ylabel(r'Temperature ($^\circ$C)')
axa.plot(pt, df[start:cutoff][['Temp [C]']].to_numpy(), 'r--')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
# plt.savefig('plots/ratetemperature.png')
# print(binned)
plt.show()

plt.clf()
fig, ax = plt.subplots()
axa = ax.twinx()
fig.set_size_inches(12, 4)
fig.set_dpi(100)
ax.set_xlabel('Timestamp (s)')
ax.plot(pt, ps)
ax.set_ylabel('Pressure (Pa)')
ax.set_ylim(ax.get_ylim()[-1]+200, ax.get_ylim()[0]+200)
axa.set_ylabel('Altitude (m)')
axa.plot(dt, alt, 'r--')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
# plt.savefig('plots/pressurealtitude.png')
# print(binned)
plt.show()

plt.clf()
fig, ax = plt.subplots()
axa = ax.twinx()
fig.set_size_inches(12, 4)
fig.set_dpi(100)
ax.set_xlabel('Flight progress')
ax.set_ylabel('Rate (Hz)')
b = [binned[0][i]/(wbin - sumdead[i]) for i in range(nbin)]
berr = [np.sqrt(binned[0][i])/(wbin - sumdead[i]) for i in range(nbin)]
xbinn = [(a-times[0])/(times[-1]-times[0]) for a in xbin]
ax.plot(xbinn, b)
ax.fill_between(xbinn, [b[i]-berr[i] for i in range(nbin)], [b[i]+berr[i] for i in range(nbin)], color=[0.5,0.5,0.5], alpha=0.25, linewidth=0)
# ax.errorbar(xbin, b, yerr=berr, color='black', fmt='.', ecolor='black', capsize=3, elinewidth=1.8)
axa.set_ylabel('Altitude (m)')
axa.plot([(a-times[0])/(times[-1]-times[0]) for a in dt], alt, 'r--')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
plt.savefig('plots/normaltime.png')
```

```python
from matplotlib.collections import LineCollection
```

```python
lat = [float(geodata['coord'][i].split(' ')[0]) for i in range(0, len(geodata['coord']), 10)]
long = [float(geodata['coord'][i].split(' ')[1]) for i in range(0, len(geodata['coord']), 10)]
color = [np.abs(i/len(lat)) for i in range(len(lat)-1)]
color = np.asarray(color)
# print(color)
norm = plt.Normalize(0, 1)
# Create a set of line segments so that we can color them individually
# This creates the points as a N x 1 x 2 array so that we can stack points
# together easily to get the segments. The segments array for line collection
# needs to be (numlines) x (points per line) x 2 (for x and y)
points = np.array([lat, long]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
# print(segments)

plt.clf()
fig, ax = plt.subplots()
fig.set_size_inches(6.5, 6)
fig.set_dpi(100)
plt.grid(visible=True)

# Create a continuous norm to map from data points to colors
lc = LineCollection(segments, cmap='twilight', norm=norm)
# Set the values used for colormapping
lc.set_array(color)
lc.set_linewidth(2)
line = ax.add_collection(lc)
cb = fig.colorbar(line, ax=ax)
cb.set_label('Flight progress')
ax.set_xlim(-70.97, -70.68)
ax.set_ylim(42.53, 42.87)
ticks = [round(-70.95+0.05*j,2) for j in range(6)]
plt.xticks(ticks, labels=['{j:.2f}'.format(j=t) for t in ticks])
plt.xlabel(r'Longitude ($^\circ$W)')
plt.ylabel(r'Latitude ($^\circ$N)')
plt.text(lat[0]-0.005, long[0]-0.005, r'$\bigstar$')
plt.text(lat[0]-0.01, long[0]-0.03, 'BVY')
plt.savefig('plots/tracker.png')
```

```python
from scipy.stats import linregress
```

```python
xt = [10*i for i in range(0,int(dt[-1]-dt[0]),10)]
mpress = [0 for i in range(0,int(dt[-1]-dt[0]),10)]
i = 0
c = 0
for j in range(len(ps)):
    if pt[j]-dt[0] > (i+1)*10:
        if c > 0:
            mpress[i] /= c
            c = 0
        i += 1
    mpress[i] += ps[j][0]
    c += 1 
mpress[-1] /= c
# print(mpress)
alts = [0 for i in range(0,int(dt[-1]-dt[0]),10)]
i = 0
c = 0
for j in range(len(alt)):
    if dt[j]-dt[0] > (i+1)*10:
        if c > 0:
            alts[i] /= c
            c = 0
        i += 1
    alts[i] += alt[j]
    c += 1 
alts[-1] /= c
# print(mpress)

res = linregress(alts, mpress)
print(f"R: {res.rvalue:.6f}")
print(f"R-squared: {res.rvalue**2:.6f}")
print(res.intercept, res.slope)

plt.clf()
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
fig.set_dpi(100)
# ax.set_xlabel('Timestamp (s)')
ax.scatter(alts, mpress, marker=',', linewidths=None, s=1, c=[[0,0,0,1] for i in range(len(alts))])
ax.plot(alts, [res.intercept + res.slope*a for a in alts], 'r--')
plt.text(500, 98000, r'$P$ = ' + str(round(res.intercept)) + r' - ' + str(abs(round(res.slope))) + r'$z$')
plt.text(500, 99000, r'$R^2 = 0.996$')
ax.set_ylabel('Pressure (Pa)')
ax.set_ylim(104000, 90000)
ax.set_xlabel('Altitude (m)')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
plt.savefig('plots/pressurealtituderegress.png',bbox_inches='tight',pad_inches=0.1)
# print(binned)
# plt.show()

# plt.clf()
# fig, ax = plt.subplots()
# axa = ax.twinx()
# fig.set_size_inches(12, 4)
# fig.set_dpi(100)
# ax.set_xlabel('Timestamp (s)')
# ax.plot(xt, mpress)
# ax.set_ylabel('Pressure (Pa)')
# ax.set_ylim(ax.get_ylim()[-1]+200, ax.get_ylim()[0]+200)
# axa.set_ylabel('Altitude (m)')
# axa.plot(xt, alts, 'r--')
# # plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
# #plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
# # plt.savefig('plots/pressurealtitude.png')
# # print(binned)
# plt.show()
```
