```python
# import pandas as pd
import math
import numpy as np
# import scipy
import matplotlib.pyplot as plt
```

```python
from cycler import cycler
plt.rcParams.update({'font.size': 14})
#plt.rcParams.update({'font.family': 'cmr10', 'mathtext.fontset': 'cm'})
plt.rcParams.update({'font.family': 'cmss10', 'mathtext.fontset': 'stixsans'})
plt.rcParams['axes.prop_cycle'] = cycler(color='kbgrcmy')
```

```python
run = 1
which = 'top'
```

```python
df = pd.read_csv('f'+str(run)+'/'+which+'.csv')
```

```python
events = df[['Event']].to_numpy()
timestamp = df[['TimeStamp [ms]']].to_numpy()
plt.clf()
plt.plot(events, timestamp)
plt.show()
```

```python
cutoff=-1
timestamp = df[:cutoff][['TimeStamp [ms]']].to_numpy()
coin = df[:cutoff][['Coincident']].to_numpy()
valid = np.nonzero(coin)
times = [timestamp[i][0]/1000 for i in valid[0]]
```

```python
plt.clf()
fig, ax = plt.subplots()
fig.set_size_inches(8, 4)
fig.set_dpi(100)
plt.xlabel('Timestamp (s)')
plt.ylabel('Count')
plt.hist(times, bins=40, edgecolor='black', linewidth=1.2, histtype='step')
# plt.hist(times, bins=50, edgecolor='black', linewidth=1.2, histtype='step')
#plt.bar(nonempty, counts, width=1, color=(1,1,1,0), yerr=err)
```
